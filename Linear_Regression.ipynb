```python
import pandas as pd
data=pd.read_csv('Salary_dataset.csv')
data.head(11)
```

```python
x=data['YearsExperience']
y=data['Salary']
x=x.fillna(x.mean())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train.values.reshape(-1, 1),y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
y_pred=model.predict(x_test.values.reshape(-1, 1))
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R*R Score:{r2}')
```

```python
value1 = float(input("Enter the years of experience: "))
new_data=[[value1]]
prediction=model.predict(new_data)
print(f'Predicted Salary for {value1} years of experience is {prediction[0]}')
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train,color='red',label='Actual')
plt.plot(x_train,model.predict(x_train.values.reshape(-1, 1)),color='blue',label='Predicted')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression for the Salary Dataset')
plt.legend()
plt.show()
```

```python
print("Training set score:{:.3f}".format(model.score(x_train.values.reshape(-1, 1),y_train)))
print("Test set score:{:.3f}".format(model.score(x_test.values.reshape(-1, 1),y_test)))
```
